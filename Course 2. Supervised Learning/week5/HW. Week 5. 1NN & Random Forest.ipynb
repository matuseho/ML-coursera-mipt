```python
import pandas as pd
import numpy as np
from sklearn import datasets, metrics, model_selection, tree, ensemble
```

```python
X, y = datasets.load_digits(return_X_y = True)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, shuffle = False, test_size = 0.25)
```

```python
X_train[0]
```

```python
def oneNN(X_train,y_train, X_test, y_test):
    y_pred = []
    for i_elem in range(len(X_test)):
        min_dist = sum((X_test[i_elem] - X_train[0])**2)
        i_min = 0
        
        for i_pair in range(len(X_train)):
            dist = sum((X_test[i_elem] - X_train[i_pair])**2)
            if dist < min_dist:
                print(dist, i_pair, y_train[i_pair])
                min_dist = dist
                i_min = i_pair
        prediction = y_train[i_min]
        print('pred:', prediction)
        y_pred.append(prediction)
    return y_pred
```

```python
y_pred = oneNN(X_train,y_train, X_test, y_test)
```

```python
acc1 = (y_pred == y_test).sum() / len(y_test)
```

```python
with open("1nn_ans1.txt", "w") as fout:
        fout.write(str(1 - acc1))
```

```python
model = ensemble.RandomForestClassifier(n_estimators=1000)
model.fit(X_train,  y_train)
y_pred = model.predict(X_test)
acc2 = metrics.accuracy_score(y_test, y_pred)
```
